### Representing a sine wave
A=sine(wt)
A=Amplitude
w=frequency
t=time


```python
import numpy as np
import matplotlib.pyplot as pl
Time=np.arange(1,10,0.1)
Amplitude=np.sin(Time)
pl.grid()
pl.plot(Time,Amplitude)
pl.title("Sine wave A=sin(Wt)")
pl.xlabel('Time')
pl.ylabel('Amplitude')
pl.show()
```

```python
import pandas as pd
df=pd.read_csv('income.csv')
df
```

```python
import matplotlib.pyplot as pl
years=df.columns[2::]
Alaska_income=df.iloc[1,2:]
pl.plot(years,Alaska_income)
pl.title("Income of Alaska from 2005-2013")
pl.xlabel("years")
pl.ylabel("Income of Alaska")
pl.savefig("Income of Alaska from 2005-2013.jpg")
pl.show()
```

```python
# plot for all states income from 2005-2013
## x-axis=years
## y-axis=All states incomes
import matplotlib.pyplot as pl
import pandas as pd
years=df.columns[-5::]
allstate_income=df.iloc[0:,2:]
allstate_income
pl.plot(years,allstate_income)
pl.title("All State income from 2009-2013")
pl.xlabel("years")
pl.ylabel("Income")
pl.savefig('All State income from 2009-2013.jpg')
pl.show()
```

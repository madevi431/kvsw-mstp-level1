### Ridge Regression Lasso Regression Elastic net

```python
from sklearn.datasets import load_boston
```

```python
boston=load_boston()
```

```python
boston
```

```python
boston.data
```

```python
boston.target
```

```python
boston.feature_names
```

```python
#creating dataset
import pandas as pd
boston_df=pd.DataFrame(boston.data,columns=boston.feature_names)
```

```python
boston_df.head()
```

```python
boston_df["Land Rate"]=boston.target
```

```python
boston_df.head()
```

```python
x=boston_df.drop("Land Rate",axis=1)
x.values
```

```python
y=boston_df["Land Rate"]
y.values
```

```python
#Ridge Regression=sumof error+sum of sqres of magnitude of 
 #coeffient(alpha)
from sklearn.linear_model import Ridge
rr=Ridge(alpha=1,solver='svd')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,
                                               random_state=1)
```

```python
rr.fit(x_train,y_train)
#print(x_train)
```

```python
y_predict=rr.predict(x_train)
```

```python
y_predict
```

```python
#score
rr.score(x_train,y_train)
```

```python
rr.score(x_test,y_test)
```

```python
rr.coef_
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lrtrain_score=lr.score(x_train,y_train)
print("LR Train Score",lrtrain_score)
```

```python
lrtest_score=lr.score(x_test,y_test)
print("LR Testing Score",lrtest_score)
```

```python
lr.coef_
```

```python
import matplotlib.pyplot as plt
plt.plot(rr.coef_,marker="*",label="Ridge coefficient",color="r")
plt.plot(lr.coef_,marker=".",label="Linear Coeffi",color="y")
plt.xlabel("coefficient index")
plt.ylabel("coefficient magnitude")
plt.title("Ridge Vs Linear")
plt.legend()
plt.show()
```

```python
rr2=Ridge(alpha=100,solver="cholesky")
```

```python
rr2.fit(x_train,y_train)
```

```python
rr2.coef_
```

```python
plt.plot(rr2.coef_,marker="*",label="Ridge coefficient",color="r")
plt.plot(lr.coef_,marker=".",label="Linear Coeffi",color="y")
plt.xlabel("coefficient index")
plt.ylabel("coefficient magnitude")
plt.title("Ridge Vs Linear")
plt.legend()
plt.show()
```

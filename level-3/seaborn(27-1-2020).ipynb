```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
sns.set(style="darkgrid")
t=sns.load_dataset("tips")
t.tail()
```

```python
sns.relplot(x="total_bill",y="day",data=t)
```

```python
sns.relplot(x="total_bill",y="day",data=t,hue="tip")
```

```python
sns.relplot(x="total_bill",y="size",data=t,hue="tip",kind="line")
```

```python
sns.set(style="white")
f=sns.load_dataset("flights")
f.head()
```

```python
sns.relplot(x="year",y="passengers",hue="month",data=f)
```

```python
sns.relplot(x="year",y="passengers",hue="month",col="month",data=f)
```

```python
sns.set(style="whitegrid")
i=sns.load_dataset("iris")
i.head()
```

```python
i["species"].value_counts()
```

```python
sns.relplot(x="sepal_width",y="petal_width",data=i)
```

```python
sns.relplot(x="sepal_width",y="petal_width",data=i,
            hue="species",col="species")
```

```python
sns.relplot("total_bill","tip",data=t,
            style="smoker",size="size",
            col="time",legend="full")
```

```python
#linear regression model plot
sns.lmplot("total_bill","tip",hue="smoker",data=t,col="time")
```

```python
#categorical data plot
#catplot
    #SWARM PLOT
    #SCATTER PLOT
    #BAR PLOT
    #violin plot
```

```python
sns.catplot(x="species",y="petal_length",data=i)
```

```python
sns.catplot(x="species",y="petal_length",
            hue="sepal_width",data=i)
```

```python
sns.catplot(x="species",y="petal_length",
            hue="sepal_width",data=i,kind="swarm")
```

```python
sns.catplot(x="species",y="petal_length",
            data=i,kind="violin")
```

```python
#box plot
f,axs=plt.subplots(1,2)
sns.boxplot(x="species",y="sepal_length",data=i,ax=axs[0])
sns.scatterplot(x="species",y="sepal_length",
                hue="petal_length",data=i,ax=axs[1])
plt.show()
```
